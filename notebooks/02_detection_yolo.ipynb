## Переименование фалов с разметкой

```python
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()
DATASET_DIR = PROJECT_ROOT / "dataset"

for split in ["train", "val"]:
    img_dir = DATASET_DIR / "images" / split
    lbl_dir = DATASET_DIR / "labels" / split

    image_stems = {p.stem for p in img_dir.glob("*") if p.suffix.lower() in {".jpg", ".jpeg", ".png"}}

    for label_path in lbl_dir.glob("*.txt"):
        label_stem = label_path.stem

        if label_stem in image_stems:
            continue

        candidate = label_stem.replace("__", " _")

        if candidate in image_stems:
            new_path = label_path.with_name(candidate + ".txt")

            if new_path.exists():
                print(f"SKIP: {new_path.name} already exists")
                continue

            label_path.rename(new_path)
            print(f"{label_path.name} -> {new_path.name}")
```

### Проверка соответствия

```python
PROJECT_ROOT = Path("..").resolve()
DATASET_DIR = PROJECT_ROOT / "dataset"

for split in ["train", "val"]:
    img_dir = DATASET_DIR / "images" / split
    lbl_dir = DATASET_DIR / "labels" / split

    imgs = {p.stem for p in img_dir.glob("*") if p.suffix.lower() in {".jpg", ".jpeg", ".png"}}
    labels = {p.stem for p in lbl_dir.glob("*.txt")}

    print(f"\n[{split}]")
    print("images:", len(imgs))
    print("labels:", len(labels))
    print("без labels:", len(imgs - labels))
    print("лишние labels:", len(labels - imgs))

    if imgs - labels:
        print("Примеры без labels:", list(sorted(imgs - labels))[:10])

    if labels - imgs:
        print("Примеры лишних labels:", list(sorted(labels - imgs))[:10])
```
